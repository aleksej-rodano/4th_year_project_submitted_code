### Slide 1: 

What are TN and Why TN?
* Tensor Networks allow us to decompose high-dimensional data into smaller, interconnected tensors. We visualize this using Penrose notation, where shapes are tensors and legs are indices.

When we connect these tensors in a 1D chain, we get a Matrix Product State (or MPS).

**There are mainly 2 advantages in usung TN:**
1.  They can be used to compress large language models.
2.  Can be easily implemented on quantum circuits.

Currently classifiers with TN have not lived up to their promises, as they have low accuracy. To improve these models, we need a better understanding of how the data is prepared.

### Slide 2
I applied the 'Quantics encoder' to the MNIST dataset.

The most important property of this encoder is that it maps images into an MPS, in such a way that different scales features are encoded into different sites along the chain.


Once we encoded all the images, we can you then create a classification, which is done by generating a sum state for each class and compress them via SVD, this allows to retain only key features.

### Slide 3
So far I have made two main investigations:

My first was on creating masks.
This is done by selecting each label in te classifier and then contracting all the shared indices, this creates a mask for each class.

This result is interesting as it is very similar to a more standard approach to create masks.

**The method is:**
Take the average of all the images in class $i$ and subtract the average of all the images that are not in class $i$.

As you can see they resemble each other. 

This is a nice property that our TN classifiers naturally captures statistical features.
